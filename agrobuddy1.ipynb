```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt
```

```python
uploaded = files.upload()
for fn in uploaded.keys():
    img_path = fn
    img = image.load_img(img_path, target_size=(224, 224))  # Resize to model input size
    plt.imshow(img)
    plt.axis('off')
    plt.title("Uploaded Crop Image")
    plt.show()
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = tf.keras.applications.mobilenet_v2.preprocess_input(x)
    model = tf.keras.applications.MobileNetV2(weights='imagenet')
    preds = model.predict(x)
    decoded = tf.keras.applications.mobilenet_v2.decode_predictions(preds, top=3)[0]
    print("\nðŸŒ¿ **Predictions:**")
    for i, (id, label, score) in enumerate(decoded):
        print(f"{i+1}. {label} â€” {score*100:.2f}%")
    disease_classes = [
        "Tomato Leaf Blight",
        "Tomato Mosaic Virus",
        "Healthy Tomato Leaf",
        "Potato Early Blight",
        "Potato Late Blight"
    ]
    print("\nâœ… Example output:")
    print(f"Predicted Disease: {np.random.choice(disease_classes)}")
```
